# Matplotlib을 이용한 보간

```python
im = mpimg.imread("../images/lena_small.jpg")                                # ndarray 영상 읽기
methods = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'lanczos'] # 보간 방법들

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10), subplot_kw={'xticks': [], 'yticks': []}) # 행,열,크기 정하기 , 그래프에 축 없음
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(im, interpolation=interp_method) # 보간 방법들 대입
    ax.set_title(str(interp_method), size=20) # 글씨 출력 및 크기 설정
plt.tight_layout() # 여백 생성
plt.show() # 영상 표시
```

# scikit-image를 사용하여 영상 읽기

H(Hue : 색상) 색상은 0~360도로 빨강(0)을 시작으로 변하게 된다.

S(Saturation : 채도) 채도는 색상의 진함을 뜻하며, 0% ~ 100%로 100%일 경우 색이 진함을 뜻한다.

V(Value : 명도) 명도는 밝고 어두움을 뜻하고, 검은색이 0% ~ 흰색이 100% 이다.

HSV를 사용하는 이유는,색상값만 있는 RGB보다 채도와 명도가 같이 있는 HSV를 사용했을 때 빛의 영향에 맞춰서 색상 기준을 조절할 수 있기 때문이다.

```python
im = imread("../images/iron_man2.png")            # 영상 읽기
print(im.shape, im.dtype, type(im))

hsv = color.rgb2hsv(im)                         # 컬러 공간 변환 (RGB --> HSV)
hsv[:, :, 1] = 0.5                              # 채도 채널 값 변경
im1 = color.hsv2rgb(hsv)                        # 컬러 공간 환원(HSV --> RGB)

imsave('../images/iron_man2_hsv.png', im1)        # 영상 저장
im = imread("../images/iron_man2_hsv.png")        # 영상 읽기
plt.axis('off')                                 # 그래프에 축 없음
imshow(im)                                      # 영상 표시
```
